# OneHot Representation and Text Preprocessing


```python
messages=x.copy()
```

```python
messages
```

```python
messages.reset_index(inplace=True)
```

```python
messages['title'][6]
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger') 
```

```python
import re
from nltk import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
from nltk import sent_tokenize,word_tokenize

```

```python
ps=PorterStemmer()
```

```python
corpus1=[]
for i in range(0,len(messages)):
  reg=re.sub('[^a-zA-Z]',' ',messages['title'][i])
  reg=reg.lower()
  reg=reg.split()
  
  rev=[ps.stem(word)for word in reg if word not in set(stopwords.words('english'))]
  rev=' '.join(rev)
  corpus1.append(rev)
```

```python
corpus1
```

```python
onehot_rep=[one_hot(word,vocabsize) for word in corpus1]
```

```python
onehot_rep
```

# **Embedding**

```python
length=20
pad=pad_sequences(onehot_rep,maxlen=length,padding='pre')
```
